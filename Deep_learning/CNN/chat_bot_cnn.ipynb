```python
import numpy as np
import pandas as pd
import tensorflow as tf

# 전처리에 필요한 라이브러리
from tensorflow.keras import preprocessing

# 함수형 방식으로 신경망 구축을 위한 라이브러리
from tensorflow.keras.layers import Input, Embedding, Dropout, Conv1D, GlobalMaxPooling1D, concatenate, Dense

# 각 레이어별 변수들을 하나로 묶어 신경망을 만들어 주는 라이브러리
from tensorflow.keras.models import Model
```

```python
#데이터 분석

data = pd.read_csv('/content/chatbot_data.csv', delimiter=',')
data.info()
```

```python
features = data['Q'].tolist() # 독립변수(질문)
labels = data['label'].tolist() # 종속변수(감정 - 사랑/이별/일상다반사)

print(features[:5])
print(labels[:5])
```

```python
# 전체 문장을 각각의 문장으로 나눠주고,
# 각각의 문장에 있는 단어별로 나눠주는 과정.

corpus_row = []
for text in features :
  corpus = preprocessing.text.text_to_word_sequence(text)
  corpus_row.append(corpus)

corpus_row
```

```python
# 텍스트를 단어 기반으로 토큰화 시켜주는 인스턴스 변수를 생성하는 과정
tokenizer = preprocessing.text.Tokenizer()
```

```python
# 위의 그림에 맞춰 코드 작성
tokenizer.fit_on_texts(corpus_row) #토큰을 시퀀스 번호로 변환해주는 과정
sequences = tokenizer.texts_to_sequences(corpus_row) # 단어 -> 시퀀스 번호로 바뀐 것을 바탕으로 단어 임메딩 벡터 생성과정

print(sequences)
```

```python
pd.DataFrame(sequences)
# 가장 큰 벡터값을 찾는 과정
#NaN을 0으로 채워주어야 합니다.
```

```python
# 패딩처리 과정
padded_seqs = preprocessing.sequence.pad_sequences(sequences, maxlen=15, padding='post')
pd.DataFrame(padded_seqs)
#pre는 앞부분으로 패딩처리 post는 뒷부분으로 패딩처리
```

# 1. 독립변수에 따른 학습 데이터 및 테스트 데이터 분리 과정

```python
shuffle_map = np.arange(len(padded_seqs), dtype='int')
print("shuffle_map : ", shuffle_map)

np.random.shuffle(shuffle_map)

print("shuffle_map : ", shuffle_map)

train_ratio = 0.7 

test_begin_index = int(padded_seqs.shape[0] * train_ratio)
print("test_begin_index : ",test_begin_index)

train_x = padded_seqs[shuffle_map][:test_begin_index, :]
test_x = padded_seqs[shuffle_map][test_begin_index:, :]
print(train_x.shape)
print(test_x.shape)
```

# 2. 종속변수에 따른 학습 데이터 및 테스트 데이터 분리 과정

```python
shuffle_map_list = shuffle_map.tolist()

label_y_row = []
for i in shuffle_map:
    label_y = labels[i]
    label_y_row.append(label_y)

label_y_row = np.asarray(label_y_row)

train_y = label_y_row[:test_begin_index]
test_y = label_y_row[test_begin_index:]

print(train_y.shape)
print(test_y.shape)
```

image.png

### 1. 단어 임베딩

```python
# 각각을 변수화 처리 해주어야 합니다.

input_layer = Input(shape = (15,))
# 패딩 처리의 max값으로 shape을 설정

embedding_layer = Embedding(input_dim=len(word_index)+1, output_dim=128, input_length=15)(input_layer)
#input_dim은 단어의 개수(카운팅이 1부터 되어 있기 때문에)
#output_dim은 Embedding 사이즈가 몇으로 되어 있는지!
#input_length 몇 개의 크기 -> 15개의 독립변수
#함수형 계층으로 신경망을 구축할 경우 이전 계층이 어떤 계층인지 명시를 해주어야 함!

dropout_emb = Dropout(rate = 0.5)(embedding_layer)
# Drop 시킬 분포만 결정
```

## 2. 특징 추출

image.png

전체 범위에 대한 maxpooling -> 대표값 하나를 뽑아주는 과정
<br>추출한 각 값을 하나로 합쳐주는 과정
<br>1차원 백터이기 때문에 완전 연결 계층에 던져주면 선택분류 가능!

```python
conv1 = Conv1D(filters=128, kernel_size=3, padding='valid',activation='relu')(dropout_emb)
#valid => 패딩을 주지 않겠다.
pool1 = GlobalMaxPooling1D()(conv1)
# 기본값 사용

conv2 = Conv1D(filters=128, kernel_size=4, padding='valid',activation='relu')(dropout_emb)
pool2 = GlobalMaxPooling1D()(conv2)

conv3 = Conv1D(filters=128, kernel_size=5, padding='valid',activation='relu')(dropout_emb)
pool3 = GlobalMaxPooling1D()(conv3)

# 하나의 행만 주어져있기 때문에 칸수로 kernel 사이즈로 가져간다!

concat = concatenate([pool1, pool2, pool3])
# 하나로 합쳐주는 과정
```

# 완전연결 계층과 연결해주는 과정

image.png

```python
hidden = Dense(128, activation='relu')(concat)

# 성능의 향상을 위해 완전 연결 계층을 하나 더 추가 합니다.
dense_2 = Dense(64, activation='relu')(hidden)

dropout_hidden = Dropout(rate=0.5)(dense_2)
logits = Dense(3, name='logits')(dropout_hidden)
# 세가지로 축약
predictions = Dense(3, activation='softmax')(logits)
# 예측
```

# 변수를 하나로 묶어주는 과정

```python
model = Model(inputs= input_layer , outputs = predictions)
#입력계층과 출력계층 지정
model.summary()
# 함수형 방식으로 신경망을 구축하는 과정
# 병렬 형식으로 만들어줄 때 함수형 방식을 활용
```

# complie

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001), 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

# 학습

```python
history = model.fit(train_x, train_y,    
                    batch_size = 16,
                    validation_split = 0.3, 
                    callbacks = [tf.keras.callbacks.EarlyStopping(patience=7)],
                    epochs=50)
```

#### 결론 : 자연어를 합성곱 신경망으로 푸는 과정
input_layer에서 맞지 않는 과정을 해결하기 위해 padding 과정을 진행!

# 시각화
